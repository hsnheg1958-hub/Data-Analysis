```python
from mpl_toolkits import mplot3d
plt.figure()
ax = plt.axes(projection='3d')
x=np.arange(-4,4,0.25)
y=np.arange(-4,4,0.25)
x,y=np.meshgrid(x,y)
R=np.sqrt(x**2+y**2)
Z=np.sin(R)
ax.plot_surface(x,y,Z)
plt.show()


```
