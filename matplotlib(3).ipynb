```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x=np.arange(-4,4,0.1) # x value
y=np.arange(-4,4,0.1) #y value
X,Y=np.meshgrid(x,y) # combine them.
R=np.sqrt(X**2+Y**2) # function
Z=np.sin(R) # Z value
ax.plot_surface(X,Y,Z, cmap='hot', edgecolors='r')
ax.set_xlabel('X-Axsis')
ax.set_ylabel('Y-Axsis')
ax.set_zlabel('Z-Axsis')
```

```python
from mpl_toolkits.mplot3d import Axes3D
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
r=np.linspace(0,6,20)   
theta=np.linspace(-0.9*np.pi,0.7*np.pi,40) # This code is to cut the graph to see its inside.
r,theta=np.meshgrid(r,theta)
X=r*np.sin(theta)
Y=r*np.cos(theta)
Z=f(X,Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='Reds_r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
