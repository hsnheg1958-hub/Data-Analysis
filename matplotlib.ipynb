```python
x=np.arange(0,10,0.005)
def p1(x):
    return x**4-4*x**3+10
def p2(x):
    return 10*x**3-48*x**2+100
def p3(x):
    return 20*x**2-96*x+100
plt.plot(x,p1(x),'--',label='p1')
plt.plot(x,p2(x),label='p2')
plt.plot(x,p3(x),label='p3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Polynomial Functions')
plt.legend()
plt.show()
```

```python
n=12
x=np.arange(n)
y1=(1-x/float(n))*np.random.uniform(0.5,1.0,n)
y2=(1-x/float(n))*np.random.uniform(0.5,1.0,n)
plt.bar(x,y1,facecolor='cyan',edgecolor='k',linewidth=1)
plt.bar(x,-y2,facecolor='r',edgecolor='k',linewidth=1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot with Positive and Negative Values') 
for x,y1 in zip(x,y1):
    plt.text(x,y1,'%.2f' % y1, ha='center', va='bottom', fontsize=10, color='black')
for x,y2 in zip(x,y2):
    plt.text(x,-y2,'%.2f' % y2, ha='center', va='top', fontsize=10, color='r')
plt.show()
```

```python
x=np.random.randint(1,10,6)
y=np.random.randint(1,10,6)
z=np.random.randint(1,10,36)
zz=np.reshape(z,(6,6))
#The `reshape` function in NumPy is used to change the shape of an array without changing its data. In this case, the `reshape` function is reshaping the array `z` into a 6x6 matrix and storing it in the variable `zz`. This allows the data to be plotted as a contour plot with the x and y values provided by the arrays `x` and `y`.

plt.contour(x,y,zz)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Contour Plot')
plt.show()
```

```python
def f(x,y):
    e=np.sin(x)**10 + np.cos(10+y*x)
    return e
x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
X,Y=np.meshgrid(x,y)
#`meshgrid` is a function in NumPy that creates a grid of coordinates for given 1D arrays. It takes two 1D arrays representing the x and y coordinates and returns two 2D arrays representing the x and y coordinates of all the points in the grid. This is commonly used in creating 2D plots and surfaces in order to evaluate functions over a grid of points.
Z=f(X,Y)
plt.contour(X,Y,Z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Contour Plot')
plt.show()
```
