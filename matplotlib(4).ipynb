# 3D:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# Create figure an 3D axis:

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
# Generate sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
# Plot surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot - sin(sqrt(x² + y²))')
fig.colorbar(surf, ax=ax, shrink=0.5)
plt.show()
```

# 3D Scatter Plot Example

```python
fig2 = plt.figure(figsize=(10, 8))
ax2 = fig2.add_subplot(111, projection='3d')
np.random.seed(42)
n = 100
xs = np.random.standard_normal(n)
ys = np.random.standard_normal(n)
zs = np.random.standard_normal(n)
colors = np.random.rand(n)
scatter = ax2.scatter(xs, ys, zs, c=colors, marker='o', s=50, cmap='plasma')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_zlabel('Z axis')
ax2.set_title('3D Scatter Plot')
fig2.colorbar(scatter, ax=ax2)
plt.show()
```

```python
x=np.linspace(0,100)
grid=plt.GridSpec(3,3)
# this code is to create a grid for the graph. I can write plt.GridSpec(3,3) or plt.GridSpec(3,3, wspace=0.2, hspace=0.2) to create a grid with space between the graphs.
plt.subplot(grid[0,0])
plt.plot(x,np.sin(x))
plt.title('sin(x)')
plt.subplot(grid[0,1])
plt.plot(x,np.cos(x))
plt.title('cos(x)')
plt.subplot(grid[0,2])
plt.plot(x,np.tan(x))
plt.title('tan(x)')
plt.subplot(grid[1,0:])
plt.plot(x,np.tan(x))
plt.title('tan(x)')
plt.subplot(grid[2,0])
plt.plot(x,2*np.cos(x))
plt.title('2 cos(x)') 
plt.subplot(grid[2,1:])
plt.plot(x,np.exp(x))
plt.title('exp(x)') 
plt.subplots_adjust(left=0, right=1.5,bottom=0, top=1.5, wspace=0.2,hspace=0.2)# this code is to determine dimentions for the graph/
plt.show()
```

# Polar:

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111, polar=True)
r=np.arange(0,1,0.001)
theta=np.pi*r*4
line=ax.plot(theta,r,c='r', label='r = 4θ',lw=2)
#` this code is to create a polar plot. I can write ax.plot(theta,r,c='r', label='r = 4θ',lw=4) to create a polar plot with label and line width.`
ax.legend()
plt.show()
```
