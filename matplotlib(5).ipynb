# Annotate الأسهم 

```python
fig,ax=plt.subplots()
x=np.linspace(0,20,100)
y=-(x-10)**2+5
ax.plot(x,y,c='r')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.annotate('Max', xy=(10, 6), xytext=(15, 10),
             arrowprops=dict(facecolor='r'))
ax.set_title('Parabola')
plt.annotate('5', xy=(5, -20), xytext=(2, 0),
             arrowprops=dict(facecolor='b', shrink=0.05,width=3, headwidth=6)) 
plt.annotate('15', xy=(15, -20), xytext=(10, -60),  
             arrowprops=dict(facecolor='g', shrink=3))
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
fg,ax=plt.subplots()
x=np.linspace(0,20,100)
ax.plot(x,np.cos(x),c='g' )
ax.axis('equal')
ax.set_title('Cosine Wave')
plt.grid(True)
plt.annotate('Local Max', xy=(6.28, 1), xytext=(10, 4),
             arrowprops=dict(facecolor='b', shrink=0.05, width=3, headwidth=6))
plt.annotate('Local Min', xy=(5*np.pi, -1), xytext=(2, -6), arrowprops=dict(facecolor='r', shrink=0.05, width=3, headwidth=6, connectionstyle='angle3, angleA=0, angleB=-90')) # this code is to create a curved arrow
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

# Legend:

```python
x=np.linspace(0,20,1000)
fig,ax=plt.subplots()
plt.axis('equal')
ax.plot(x,np.sin(x),label='sin(x)',lw=3)
ax.plot(x,np.cos(x),c='r',label='cos(x)',lw=2,linestyle='--')
leg = ax.legend(
    loc='upper center',
    frameon=True,
    shadow=True,#this code is to add a shadow to the legend box
    fancybox=True,#this code is to make the legend box have rounded corners
    facecolor='k',#this code is to set the background color of the legend box to black
    edgecolor='black',#this code is to set the font family and size of the legend text  
    ncol=2,#this code is to set the number of columns in the legend to 2 
     
    prop={'family': 'serif', 'size': 14},# this code is to set the font family and size of the legend text
    title='Trigonometric Functions',
    title_fontsize=14
)
leg.get_title().set_fontfamily('serif') 

```

```python
plt.style.use('ggplot')
x=np.linspace(0,20,1000)
fig,ax=plt.subplots()
plt.axis('equal')
ax.plot(x,np.sin(x),c='k',label='sin(x)',lw=3)
ax.plot(x,np.cos(x),c='r',label='cos(x)',lw=2,linestyle='--')
leg = ax.legend(
    loc='upper center',
    frameon=True,
    shadow=True,#this code is to add a shadow to the legend box
    fancybox=True,#this code is to make the legend box have rounded corners
    framealpha=0.1,#this code is to set the transparency of the legend box to 0.5   
    facecolor='b',#this code is to set the background color of the legend box to black
    edgecolor='black',#this code is to set the font family and size of the legend text  
    ncol=2,#this code is to set the number of columns in the legend to 2 
    borderpad=1,#this code is to set the border padding of the legend box to 1
    prop={'family': 'serif', 'size': 14, 'style': 'italic'},# this code is to set the font family and size of the legend text
    title='Trigonometric Functions',
    title_fontsize=14
)
leg.get_title().set_fontfamily('serif')
leg.get_title().set_style('italic')
leg.get_title().set_weight('bold') 

```

```python
w=[1,2,3,4,5]
c=['c','m','y','k','orange']
l=['-','--','-.',':','-']
for i in range(len(w)):
    plt.vlines(x=w[i], ymin=0, ymax=4, color=c[i], linestyles=l[i], lw=2, label=f'Vertical Line {i+1}')
```

```python
w=['-','--','-.',':','-']
c=['c','m','y','k','orange']
l=[1,2,3,4,5]
y=[5,6,7,8]
for yval, color, style, lw in zip(y, c, w, l):
        plt.hlines(y=yval, xmin=0, xmax=20, colors=color, linestyles=style, lw=lw, label='Horizontal Lines')
```

```python
# zip(y, c, w, l) this code is to create an iterator that aggregates elements from each of the iterables y, c, w, l. It returns an iterator of tuples where the i-th tuple contains the i-th element from each of the argument sequences or iterables. The iterator stops when the shortest input iterable is exhausted.
```

# Determining  the graph limits:


```python
# The first form:
x=np.linspace(-10,10)
ax=plt.axes()
ax.set_xlim(-20,20)
ax.set_ylim(-20,20)
ax.plot(x,-x**2)
```

```python
# Python determines the limits of the x and y axis automatically based on the data being plotted. However, you can manually set the limits of the axes using the set_xlim() and set_ylim() methods of the axes object. In this code, we are setting the limits of the x-axis to be from -10 to 10 and the limits of the y-axis to be from -10 to 10. This allows us to see the entire plot of the function -x^2 without any part of it being cut off.
x=np.linspace(-10,10)
ax=plt.axes()
plt.plot(x,-x**2)
```

```python
# The second foem:
x=np.linspace(-10,10)
plt.axis([-20,10,-10,20])# this code is to set the limits of the x and y axis to be from -20 to 10 and -10 to 20 respectively    
plt.plot(x,-x**2)
# It is important to set the limits of the axes when plotting a function to ensure that the entire plot is visible and not cut off. In this code, we are setting the limits of the x-axis to be from -20 to 10 and the limits of the y-axis to be from -10 to 20. This allows us to see the entire plot of the function -x^2 without any part of it being cut off.
```

```python
# The third form:
x=np.linspace(-10,10)
ax=plt.axes()
ax.set(xlim=(-20,20), ylim=(-20,20))#this code is to set the limits of the x and y axis to be from -20 to 20
ax.plot(x,-x**2)
```

```python
x=np.linspace(-10,10)
plt.axis('equal') # this code is to set the aspect ratio of the plot to be equal, which means that one unit on the x-axis is equal to one unit on the y-axis. This is useful when plotting functions that have a specific shape, such as circles or ellipses, to ensure that they are not distorted.    
plt.plot(x,-x**2)
```

```python
# Hiding the axes:
x=np.linspace(-10,10)
ax=plt.axes()
ax.xaxis.set_visible(False) # this code is to hide the x-axis
ax.yaxis.set_visible(False) # this code is to hide the y-axis
ax.plot(x,-x**2,c='r')
ax.set_title('Parabola')
```

# Patches :

```python
import matplotlib.pyplot as plt
import matplotlib.patches  as pat
```

```python
c=pat.Circle((0.5,0.5), radius=0.3,facecolor='r', edgecolor='black', lw=2,alpha=0.5)
ax=plt.axes()
ax.set_aspect('equal')
# this code is to set the aspect ratio of the plot to be equal, which means that one unit on the x-axis is equal to one unit on the y-axis. This is useful when plotting shapes like circles to ensure that they are not distorted.
ax.add_patch(c)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title('Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Ellipse:

```python
c=pat.Ellipse((0.5,0.5), width=0.4, height=0.2, angle=30, facecolor='b', edgecolor='black', lw=2, alpha=0.5)
ax=plt.axes()
ax.set_aspect('equal')
ax.add_patch(c)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title('Ellipse')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
c=pat.Rectangle((0.2,0.2), width=0.6, height=0.2, angle=30, facecolor='g', edgecolor='black', lw=2, alpha=0.5)
ax=plt.axes()
ax.set_aspect('equal')
ax.add_patch(c)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title('Rectangle')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Arc:

```python
c=pat.Arc((0.5,0.5), width=0.6, height=0.6, angle=0, theta1=0, theta2=180, facecolor='none', edgecolor='black', lw=2, alpha=0.5)# this code is to create an arc with center at (0.5,0.5), width of 0.6, height of 0.6, angle of 0 degrees, starting angle of 0 degrees and ending angle of 180 degrees. The facecolor is set to 'none' to make the arc transparent and the edgecolor is set to 'black' to make the outline of the arc visible. The lw parameter is set to 2 to make the outline thicker and the alpha parameter is set to 0.5 to make the arc semi-transparent.
ax=plt.axes()
ax.set_aspect('equal')
ax.add_patch(c)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title('Arc')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()  
```
